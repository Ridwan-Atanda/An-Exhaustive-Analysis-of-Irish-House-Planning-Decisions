## Categorizing the Target variable into 3
### Approved
### Disapproved 
### Pending 

```python
#The Decision column has 158 unique values 
#we pick a keyword from each of those unique values and divide them into three categories
#create a new column and put each of those unique keywords into three categories 



Approved = [ 'approval', 'approve', 'conditional', 'c', 'exempt', 'extension', 'ext', 'grant', 'granted', 
            'exempted', 'granted', 'approved', 'split', 'unconditional', 'approved']
Disapproved = [ 'invalid', 'withdrawn', 'disapprove', 'cannot', 'additional', 'further', 'refuse', 'invalid', 'invalidate',
               'closed', 'rejected', 'precluded', 'r', 'refuse', 'refused', 'withdraw', 'disapproved']
Pending = [ 'additions','revisions', 'returned', 'request', 'revised', 'referred', 'seek', 'extension']

```

```python
#define a function that maps each of those keywords to their respective categories

def decide(x):
    x = x.lower()
    x = x.split(' ')
    for i in x:
        if i in Approved:
            return 'Approved' 
            break
        elif i in Disapproved:
            return 'Disapproved'
            break
        elif i in Pending:
            return 'Pending'
            break
        else:
            return 'check'
            
```

```python
df2['Final_Decision'] = df2['Decision'].apply(lambda x: decide(x))
```

### Convert the date column into Pandas date format 

```python
df3 = df2.dropna(axis=0, subset=['ReceivedDate']) 
#drop missing values in the dataframe, but axis = 0 means we drop rows with missing values 
#axis =1 means we drop columns with missing value 
                                                    
df3['ReceivedDate'] = pd.to_datetime(df2['ReceivedDate']) #we convert recieve date to python date and time format
lb_list = [] #an empty list
spring = ['03','04','05']
summer = ['06','07','08']
autumn = ['09','10','11']
winter = ['12','01','02']
          
for i in range(len(df2)):
    try:
        if str(df2['ReceivedDate'][i])[5:7] in spring:  
              lb_list.append('spring')
        elif str(df2['ReceivedDate'][i])[5:7] in summer:
              lb_list.append('summer')
        elif str(df2['ReceivedDate'][i])[5:7] in autumn:
              lb_list.append('autumn')
        elif str(df2['ReceivedDate'][i])[5:7] in winter:
              lb_list.append('winter')
    except:
        lb_list.append('Nan')
print(len(lb_list))
df2['Seasons'] = lb_list
np.unique(list(df2['Seasons'].values))
```
